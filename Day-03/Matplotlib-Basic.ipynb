# 시각화 - Matplotlib

```
import numpy as np
import pandas as pd
```

```
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

### 라인 플롯

```
plt.title("x ticks") # 제목 표시
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show() # 차트로 렌더링
```

### 스타일 지정

```
plt.title("'rs--' plot ")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')
plt.show()
```

### 그림 범위 지정

```
plt.title("x, y arange setting")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.grid()
plt.show()
```

### 틱 설정

```
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x, y tick label setting")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid()
plt.show()
```

### 여러 개의 선을 그리기

```
t = np.arange(0., 5., 0.2)
plt.title("Multi-Line")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.show()
```

```
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend plot")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
```

# 그림의 구조
### Figure / Axes / Axis 객체로 구성
- Figure : 한 개 이상의 Axes 객체를 포함, 그림이 그려지는 캔버스나 종이를 뜻함.
- Axes : 두 개 이상의 Axis 객체를 포함, 하나의 플롯
- Axis : 가로 축이나 세로 축 등의 축을 뜻함.

```
np.random.seed(0)
f1 = plt.figure(figsize=(10, 2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100))
plt.grid()
plt.show()
```

### Subplot
- 그리드 형태의 Axes 객체 생성. Figure가 행렬이고, Axes가 행렬의 원소.

```
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.grid()

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.grid()
plt.tight_layout()
plt.show()
```

### 바 차트

```
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['a', 'b', 'c']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("a b c")
plt.ylabel("frequency")
plt.show()
```

### 스템 플롯

```
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title("Stem Plot")
plt.stem(x, np.cos(x), '-.')
plt.show()
```

### 파이차트

```
plt.axis('equal')
labels = ['a', 'b', 'c', 'd']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

### 히스토그램

# Seaborn 사용한 데이터 분포 시각화

```
import seaborn as sns
```

### 1차원 분포 플롯

```
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
tips = sns.load_dataset("tips")    # 팁 데이터
flights = sns.load_dataset("flights")    # 여객운송 데이터
```

```
x = iris.petal_length.values

sns.rugplot(x)
plt.title("Iris_length_Rug Plot ")
plt.show()
```

```
sns.kdeplot(x)
plt.title("Iris_length_Kernel_Density_Plot")
plt.show()
```

```
sns.distplot(x, kde=True, rug=True)
plt.title("Iris_length_Dist_Plot")
plt.show()
```

### 카운트 플롯
- countplot(x="column_name", data=dataframe)

```
sns.countplot(x="class", data=titanic)
plt.title("Each class for passenger")
plt.show()
```

```
sns.countplot(x="day", data=tips)
plt.title("Day for tips")
plt.show()
```

# 다차원 데이터
- 분석하고자 하는 데이터가 모두 실수 값인 경우
- 분석하고자 하는 데이터가 모두 카테고리 값인 경우
- 분석하고자 하는 데이터가 모두 실수 값과 카테고리 값이 섞여 있는 경우

### 2차원 실수형 데이터
- jointplot(x="x_name", y="y_name", data=dataframe, kind='scatter')

```
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
plt.suptitle("width_length_Joint Plot", y=1.02)
plt.grid()
plt.show()
```

```
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
plt.suptitle("Width_Length_Joint Plot_Kernel Density Plot", y=1.02)
plt.grid()
plt.show()
```

### 다차원 실수형 데이터

```
sns.pairplot(iris)
plt.title("Iris Data_Pair Plot")
plt.show()
```

```
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.title("Iris Pair Plot_Hue_visualization_species")
plt.show()
```

# Pandas의 시각화 기능

```
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2018', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```
df1.plot()
plt.title("Pandas_Plot_Method_Example")
plt.xlabel("Date")
plt.ylabel("Data")
plt.show()
```
