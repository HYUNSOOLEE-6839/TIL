# Decision Tree

- 변수들로 기준을 만든 것을 통하여 샘플을 분류하고 분류된 집단의 성질을 통하여 추정하는 모형.
    - 장점 : 해석력이 높음, 직관적, 범용성
    - 단점 : 높은 변동성, 샘플에 민감할 수 있음.

### 용어 - 트리구조
- Node - 분류의 기준이 되는 변수가 위치. 이것을 기준으로 샘플을 나눔.
    - Parent node - 상대적인 개념. 상위 노드.
    - Child node - 하위 노드.
    - Root node - 상위노드가 없는 가장 위의 노드.
    - Leaf node(Tip) - 하위 노드가 없는 가장 아래의 노드. Yes나 No 등 결과값.
    - Internal node - Leaf node가 아닌 노드.

- Edge - 샘플을 분류하는 조건이 위치하는 곳.
- Depth - Root node에서 특정 노드까지 도달하기 위해 거쳐야하는 Edge의 수

### 반응 변수에 따라
- 범주형 변수 - 분류트리
- 연속형 변수 - 회귀트리

### Entropy
- 0 또는 1일 확률이 최소, 0.5일 확률이 최대가 되게 하는 함수

### Classification Tree
- Tree 조건에 따라, X가 가질 수 있는 영역을 Block으로 나누는 개념.
    - 나누어진 영역 안에 속하는 샘플의 특성을 통하여 Y를 추정.

### 의사결정나무 구축 및 시각화

- 트리 구축

```python
clf=tree.DecisionTreeClassifier()
clf=clf.fit(iris.data, iris.target)
```

- 트리의 시각화

```python
dot_data=tree.export_graphviz(clf, out_file=None,
                               feature_names=iris.feature_names,
                               class_names=iris.target_names,
                               filled=True, rounded=True,
                               special_characters=True)
graph=graphviz.Source(dot_data)
```

```python
graph
```

- 엔트로피를 활용한 트리

```python
clf2=tree.DecisionTreeClassifier(criterion='entropy')
clf2=clf2.fit(iris.data, iris.target)
```

```python
dot_data2=tree.export_graphviz(clf2, out_file=None,
                               feature_names=iris.feature_names,
                               class_names=iris.target_names,
                               filled=True, rounded=True,
                               special_characters=True)
graph2=graphviz.Source(dot_data2)
```

```python
graph2
```

### 프루닝
- overfitting이 되지 않게 적절히 끊어주는 것.

```python
clf3=tree.DecisionTreeClassifier(criterion='entropy', max_depth=2) 
# max_depth : node의 개수, 즉 가지의 개수를 제한.
```

```python
clf3=clf3.fit(iris.data, iris.target)
```

```python
dot_data3=tree.export_graphviz(clf3, out_file=None,
                               feature_names=iris.feature_names,
                               class_names=iris.target_names,
                               filled=True, rounded=True,
                               special_characters=True)
graph3=graphviz.Source(dot_data3)
```

```python
graph3
```

- Confusion Matrix 구하기

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(iris.target, clf.predict(iris.data))
```

```python
# 지니계수, 엔트로피로 criterion을 정한 것.
confusion_matrix(iris.target, clf2.predict(iris.data))
```

```python
# 프루닝한 것.
confusion_matrix(iris.target, clf3.predict(iris.data))
```

- 프루닝을 했을 때 Training 셋 조차 틀리는 경우가 있음.
- 과적합이 되어있다면 간단한 모델로 이만큼 맞춘 것도 좋은 결과라고 볼 수 있음.

### Training - Test 구분 및 confusion matrix 계산

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, stratify=iris.target, random_state=1)
```

```python
clf4 = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
clf4.fit(X_train, y_train)
```

```python
confusion_matrix(y_test, clf4.predict(X_test)) # 실제 값, 예측 값
```

### Decision regression tree
- Y 반응변수가 연속형일 때 사용

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))
```

- Regression 구축

```python
regr1 = tree.DecisionTreeRegressor(max_depth=2)
regr2 = tree.DecisionTreeRegressor(max_depth=5)
# Depth가 변함에 따라 얼마나 변화를 보이는가?
```

```python
regr1.fit(X, y)
```

```python
regr2.fit(X, y)
```

```python
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
X_test[:5]
```

```python
y_1 = regr1.predict(X_test)
y_2 = regr2.predict(X_test)
```

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolor='black',
           c='darkorange', label='data')
plt.plot(X_test, y_1, color='cornflowerblue',
        label = 'max_depth=2', linewidth=2)
plt.plot(X_test, y_2, color='yellowgreen', label='max_depth=5', linewidth=2)
plt.xlabel('data')
plt.ylabel('target')
plt.title('Decision Tree Regression')
plt.legend()
plt.show()
```

- Depth 2인 경우에는 계단이 크지 않으면서 커다란 트렌드를 따라감.



- Depth 5인 경우에는 이산치에 매우 민감하게 반응하여 복잡하게 찾음.
    > => 우리가 갖고 있는 셋을 잘 찾을 수 있을까?

```python
dot_data4 = tree.export_graphviz(regr2, out_file=None,
                                filled=True, rounded=True,
                                special_characters=True)
```

```python
# Depth가 5인 경우의 그래프
graph4 = graphviz.Source(dot_data4)
graph4
```
