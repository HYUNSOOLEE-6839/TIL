### 컬럼 설명
- Status : 개발 또는 개발상태
- Adult Mortality : 성별의 성인 사망률(인구 1000명당 15~60년 사망 확률)
- Alcohol : 알코올, 1인당 소비량(15+) 기록(순수 알코올 리터 단위)
- Percentage Expenditure : 1인당 국내총생산(%) 대비 보건비 지출
- Hepatitis B : 1세 아동 중 B형 간염(HepB) 예방접종 보장률(%)
- Measles : 홍역 - 인구 1000명당 보고된 환자 수
- Polio : 소아마비
- Total expenditure : 총 지출
- Diphtheria : 디프테리아(급성 전염병)
- thiness : 영양실조
- Income composition of resources : 자원의 소득 구성

```python
# 이상치 제거 함수
def remove_outliers(data):
    arr = []
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    mi = q1 - (1.5 * iqr)
    ma = q3 + (1.5 * iqr)
    
    for i in list(data):
        if i<mi:
            arr.append(i)
        elif i>ma:
            i=ma
            arr.append(i)
        else:
            arr.append(i)
    return arr
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
data = life_data.apply(label_encoder.fit_transform)
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

# 번외 ) 각 컬럼을 통해 개발중인 국가와 개발된 국가와의 상관관계 알아보기

```python
x = data.drop('Status', axis=1)
y = data['Status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
# LR 모델 생성 및 학습
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
```

```python
# 학습 결과 평가
from sklearn.metrics import classification_report

# pred 수행하고 classification_report() 결과 출력
pred = model_lr.predict(X_test)
print(classification_report(y_test, pred))
```

```python
# LR 모델의 coef_ 속성 plot
fig = plt.figure(figsize=(10, 15))
model_coef = pd.DataFrame(data=model_lr.coef_[0], index=x.columns, columns=['Model Coefficient'])
model_coef.sort_values(by='Model Coefficient', inplace=True, ascending=False)
plt.barh(model_coef.index, model_coef['Model Coefficient'])
plt.xticks(rotation=90)
plt.grid()
plt.show();
```
