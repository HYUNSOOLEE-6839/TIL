# 다중 선형 회귀분석
    - 변수가 여러 개인 경우..
- 다중 선형 회귀계수 추정
    - 회귀계수를 추정하는 것은 단순선형회귀분석과 동일하게 SSE를 최소화하는 방향으로 추정
    - 각각의 변수에 대해 편미분하여 회귀계수를 추정
    - 단순 선형 회귀계수와 다른 점은 앞에 조건이 붙는다는 것.
    - 광고미디어를 예시로...
    - Tv가 1단위 증가할 때 매출은 0.046 증가한다...
        > 그 때의 p-value는 매우 낮기 때문에 유의미하다고 볼 수 있다.

- 다중 선형 회귀 계수 검정
    - B.p의 검정
        - 귀무가설 : B.p = 0 (회귀계수는 0이다, 변수의 설명력이 없다.)
        - 대립가설 : B.p != 0 (회귀계수는 0이 아니다. 변수의 설명력이 있다.) 
        > 회귀계수가 0에 가깝고, p-value가 매우 높다면... 귀무가설 채택.

- 다중 선형 회귀 모델 검정
    - 귀무가설 : B.1 = B.2 ... B.p = 0 (모든 회귀계수 0, 변수 설명력 x)
    - 대립가설 : 하나의 회귀계수라도 이 아니다.(설명력 있는 변수 존재.)
        - F검정을 통해서 검정
        > 두 확률 변수 V1, V2가 서로 독립인 카이제곱 분포를 따른다고 할 때 확률 변수 F는 F 분포를 따른다고 한다. F검정과 분산분석 등에서 주로 사용됨.
    
    > 귀무가설은 기각하기 너무 쉬운 가설. 변수가 추가될 수록 기각하기 쉬워진다.
    - 1) 총 변동은 정해져 있다(바뀌지 않는다)
    - 2) 제곱합의 형태이기 때문에 변수가 추가되면 추가될 수록 자연스레 증가한다.
    - 3) 제곱합의 형태로 검정을 하는 F검정의 특성상 변수가 추가되면 자연스레 기각하기 쉬워진다. (R ** 2(R-스퀘어)도 커짐) 
        > R-스퀘어 : SSR / SST

# 단순선형회귀분석 실습해보기

```python
import os 
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
# 데이터 불러오기
boston = pd.read_csv('./data/boston_train.csv')
boston.head()
```

```python
# target이 빠진 보스턴 집값 데이터 생성.
boston_data = boston.drop(['target'], axis=1)
```

```python
target = boston[['target']]
crim = boston[['CRIM']]
rm = boston[['RM']]
lstat = boston['LSTAT']
```

```python
crim1 = sm.add_constant(crim, has_constant='add')
```

```python
model1 = sm.OLS(target, crim1)
fitted_model1 = model1.fit()
```

```python
fitted_model1.summary()
```

```python
pred1 = fitted_model1.predict(crim1)
```

```python
import matplotlib.pyplot as plt
plt.yticks(fontname = "Arial")
plt.scatter(crim, target, label="data")
plt.plot(crim, pred1, label='result')
plt.legend()
plt.show()
```

```python
plt.scatter(target, pred1)
plt.xlabel('real_value')
plt.ylabel('pred_value')
plt.show()
```

```python
fitted_model1.resid.plot()
plt.xlabel('residual_number')
plt.show()
```

```python
sum(fitted_model1.resid)
```

```python
rm1 = sm.add_constant(rm, has_constant='add')
lstat1 = sm.add_constant(lstat, has_constant='add')
```

```python
model2 = sm.OLS(target, rm1)
fitted_model2 = model2.fit()
model3 = sm.OLS(target, lstat1)
fitted_model3 = model3.fit()
```

```python
fitted_model2.summary()
```

```python
fitted_model3.summary()
```

```python
pred2 = fitted_model2.predict(rm1)
pred3 = fitted_model3.predict(lstat1)
```

```python
plt.scatter(rm, target, label="data")
plt.plot(rm, pred2, label='result')
plt.legend()
plt.show()
```

```python
plt.scatter(lstat, target, label="data")
plt.plot(lstat, pred3, label='result')
plt.legend()
plt.show()
```

```python
plt.scatter(target, pred2)
plt.xlabel('real_value')
plt.ylabel('pred_value')
plt.show()
```

```python
plt.scatter(target, pred3)
plt.xlabel('real_value')
plt.ylabel('pred_value')
plt.show()
```

# 다중선형회귀분석
### crim, rm, lstat 세 개의 변수를 통해 다중회귀적합

```python
# boston data에서 crim, rm, lstat 변수만 뽑아오기

x_data = boston[['CRIM', 'RM', 'LSTAT']]
x_data.head()
```

```python
# 상수항 추가
x_data1 = sm.add_constant(x_data, has_constant='add')
```

```python
# 회귀모델 적합
multi_model = sm.OLS(target, x_data1)
fitted_multi_model = multi_model.fit()
```

```python
# summary 함수를 통해 결과 출력
fitted_multi_model.summary()
```

### 단순선형회귀모델 계수와 비교

```python
# 단순선형회귀모델의 회귀 계수
print(fitted_model1.params)
print(fitted_model2.params)
print(fitted_model3.params)
```

```python
# 다중선형회귀모델의 회귀 계수
print(fitted_multi_model.params)
```

- 선형회귀분석은 y의 변동성을 x가 얼마나 잘 설명하느냐....
- 변수가 여러 개일 때 그 설명하는 변동성을 중복으로 각각 가져가지 못 한다.
- CRIM, RM, LSTAT 3개를 모두 더 하면 1이 넘고, multi_model의 R-squared는 0.660이다.
- LSTAT의 R-squared는 0.560이였으며 그것보다 0.1 높은 수준이 됐음.
- 결국, 세 개의 변수가 각각 설명하는 y의 변동비율이 겹친다는 말임.
- 이 회귀계수에 대한 절댓값도 위와 같이 줄어드는 것임.
    > 이를 다중공산성이라고 함.
- 각각의 p-value가 낮음. 이러한 경우에도 3개의 변수를 한꺼번에 쓸 수 있는 정도임.
- 즉, 다중공산성이 있음에도 불구하고 이 회귀모델을 쓰는 것이 나쁘지는 않다라는 해석.

### 행렬연산을 통해 beta 구하기

```python
from numpy import linalg # 행렬연산을 통해 beta 구하기 - SSE를 최소화
# (X'X) - 1X'y
```

```python
np.dot(x_data1.T, x_data1)
# 이후 위 행렬의 역행렬을 구해야함.
```

```python
# 역행렬 구하기
ba = linalg.inv(np.dot(x_data1.T, x_data1))
```

```python
# 역행렬과 x-트랜스포즈를 곱하고, y도 곱해야함.
np.dot(np.dot(ba, x_data1.T), target)
```

```python
print(fitted_multi_model.params)
```

- 마지막으로 구한 연산과 다중선형회귀모델의 회귀 계수를 비교하면 같은 값이 나옴.

```python
# y_hat 구하기
pred4 = fitted_multi_model.predict(x_data1)
```

```python
### residual plot
fitted_multi_model.resid.plot()
plt.xlabel('residual_number')
plt.show()
```

```python
fitted_model1.resid.plot(label='crim')
fitted_model2.resid.plot(label='rm')
fitted_model3.resid.plot(label='lstat')
fitted_multi_model.resid.plot(label='full')
plt.legend();
```

- full 모델이 가장 잔차가 적다.

- 변수가 많아질 수록 R-squared는 증가하고, 학습데이터 내에서는 잔차가 작아짐.

- 검증 데이터에 대해서도 확인해보아야하고 다른 좋은 변수들이 있는지도 확인해야함.
